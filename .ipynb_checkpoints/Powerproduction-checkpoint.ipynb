```python
#import pandas and read powerproduction dataset
import pandas as pd
path = ("https://raw.githubusercontent.com/ianmcloughlin/2020A-machstat-project/master/dataset/powerproduction.csv")## 
df = pd.read_csv (path)
df
```

Display pairplot to see the overall aspect of each of the variables:

## Simple linear regression

<br>

Exploration of the distribution

***

```python
# Make matplotlib show interactive plots in the notebook.
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = df['power']
y = df['speed'] 

##coef = np.polyfit(x,y,1)
##poly1d_fn = np.poly1d(coef) 
## poly1d_fn is now a function which takes in x and returns an estimate for y

#plt.plot(x,y, 'yo', x, poly1d_fn(x), '--k')
#plt.xlim(0, 5)
#plt.ylim(0, 12)
m,b = np.polyfit(x, y, 1)
print(m)
print(b)
plt.title('Regression line taking into account all the dataset', fontsize=14)
plt.xlabel('Power', fontsize=12)
plt.ylabel('Speed', fontsize=12)
plt.plot(x, y, 'yo', x, m*x+b, '--k') 
## references: https://stackoverflow.com/questions/12444716/how-do-i-set-the-figure-title-and-axes-labels-font-size-in-matplotlib
## https://stackoverflow.com/questions/6148207/linear-regression-with-matplotlib-numpy
```

By looking at the result, seems at the extreme parts of the dataset, there are some values that doesn't look are following the trend of the middle values. This affects the regression line coming out after treating all values of the dataset. 

Instead if we set specific limits at the interval power = (20,80)
plt.xlim(0, 20)
plt.ylim(0, 80)

```python
plt.xlim(20, 80)
plt.ylim(0, 20)
plt.plot(x, y, 'yo', label="Data") 
```

```python
plt.plot(x, y, 'yo', label="Data") 
plt.plot(x, m*x+b, '--k', label="Regression from the entire dataset") 
plt.plot(x, 0.12 * x + 6.0, 'r-', label=r"$4x + 6$")
plt.plot(x, 0.10 * x +  8.0, 'g-', label=r"$3x +  7$")
plt.plot(x, 0.08 * x + 8.5, 'b-', label=r"$2x + 9$")
plt.legend()
```
